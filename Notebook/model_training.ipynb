```python
import pandas as pd
```

```python
df=pd.read_csv("cleaned_data")
```

```python
df
```

```python
x=df.drop(labels=["class"],axis=1)
y=df[["class"]]


```

```python
x.shape
```

```python
y.shape
```

```python
categorical_columns=x.select_dtypes(include="object").columns
numeric_columns=x.select_dtypes(exclude="object").columns
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
#for pipeline
from sklearn.pipeline import Pipeline 
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline=Pipeline(
    steps=[
        ("imputer",SimpleImputer(strategy="median")),
        ("scaler",StandardScaler())
    ]
)

cat_pipeline=Pipeline(
    steps=[
        ("imputer",SimpleImputer(strategy="most_frequent")),
        ("OrdinalEncoder",OrdinalEncoder()),
        ("scaler",StandardScaler())
    ]
)


preprocessor=ColumnTransformer(
    [
        ("num_pipeline",num_pipeline,numeric_columns),
        ("cat_pipeline",cat_pipeline,categorical_columns)
    ]
)
```

```python
# train test split


from sklearn.model_selection import train_test_split

x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.30,random_state=42)

preprocessor.fit_transform(x_train)
```

```python
x_train=pd.DataFrame(preprocessor.fit_transform(x_train),columns=preprocessor.get_feature_names_out())
```

```python
x_test=pd.DataFrame(preprocessor.transform(x_test),columns=preprocessor.get_feature_names_out())
```

```python
categorical_columns=[]
def feature_encode(categorical_columns):
    for label,content in df.items():
        if label not in categorical_columns:
            df[label]=pd.Categorical(content).codes+1

feature_encode(categorical_columns)
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
classifier=regression=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,10,20,30,40,50],'max_iter':[100,200,300]}
```

```python
classifier_regressor=GridSearchCV(classifier,param_grid=parameter,scoring="accuracy",cv=5)
```

```python
classifier_regressor.fit(x_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
y_pred=classifier_regressor.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
